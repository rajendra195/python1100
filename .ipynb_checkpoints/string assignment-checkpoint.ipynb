# Write a program that asks the user to enter a string and prints the string in reverse order.

```python
input_str = input("Enter a string? ")
reverse_str = ""
for index in range(len(input_str) - 1, -1, -1):
    letter = input_str[index]
    reverse_str += letter
print(f"The reverse of input string '{input_str}' is '{reverse_str}'.")
```

# Write a program that asks the user to enter a string and print the number of vowels in the string.

```python
input_str = input("Enter a string: ").lower()
vowels = 0
for index in range(len(input_str)):
    letter = input_str[index]
    if letter in "aeiou":
        vowels += 1
print(f"The numbers of vowels in string {input_str} is {vowels}.")
```

# Write a program that asks the user to enter a sentence and print the sentence with each word capitalized.

```python
sentence = input("Enter a sentence: ")
capital_sentence = sentence.title()
print(f"The capitalized sentence is {capital_sentence}.")
```

# WAP that asks the user to enter a string and prints the string with all the vowels replaced by the character "o".

```python
inp_str = input("Enter a string: ").lower()
replaced_str = ""
for index in range(len(inp_str)):
    letter = inp_str[index]
    if letter in "aeiou":
        letter = letter.replace(letter, "o")
        replaced_str += letter
    else:
        replaced_str += letter
print(f"The vowels letter raplaced by 'o' in a string is {replaced_str}.")
```
